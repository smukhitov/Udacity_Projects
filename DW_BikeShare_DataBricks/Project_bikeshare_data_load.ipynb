```
df_payments = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load("/FileStore/udacity/payments.csv") \
    .withColumnRenamed("_c0","payment_id") \
    .withColumnRenamed("_c1","date")\
    .withColumnRenamed("_c2","amount")\
    .withColumnRenamed("_c3","rider_id")\

#Bronze - data is ingested into Delta fs
df_payments.write.format("delta") \
        .mode("overwrite") \
        .save("/delta/payments")
#silver
spark.sql("CREATE TABLE stg_payments using DELTA LOCATION '/delta/payments'")
```

```
df_riders = spark.read.format("csv") \
                 .option("inferSchema", "false") \
                 .option("header", "false") \
                 .option("sep", ",") \
                 .load("/FileStore/udacity/riders.csv")\
                 .withColumnRenamed("_c0","rider_id") \
                 .withColumnRenamed("_c1","firstName")\
                 .withColumnRenamed("_c2","lastName")\
                 .withColumnRenamed("_c3","address")\
                 .withColumnRenamed("_c4","birthday")\
                 .withColumnRenamed("_c5","startDate")\
                 .withColumnRenamed("_c6","endDate")\
                 .withColumnRenamed("_c7","isMember")

df_riders.write.format("delta") \
        .mode("overwrite") \
        .save("/delta/riders")

spark.sql("CREATE TABLE stg_riders using DELTA LOCATION '/delta/riders'")
```

```
df_stations = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load("/FileStore/udacity/stations.csv") \
    .withColumnRenamed("_c0","station_id") \
    .withColumnRenamed("_c1","name")\
    .withColumnRenamed("_c2","latitude")\
    .withColumnRenamed("_c3","longitude")\

df_stations.write\
    .format("delta")\
    .mode("overwrite")\
    .save("/delta/stations")

spark.sql("CREATE TABLE stg_stations using DELTA LOCATION '/delta/stations'")
```

```
df_trips = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load("/FileStore/udacity/trips.csv") \
    .withColumnRenamed("_c0","trip_id") \
    .withColumnRenamed("_c1","rideable_type")\
    .withColumnRenamed("_c2","started_at")\
    .withColumnRenamed("_c3","ended_at")\
    .withColumnRenamed("_c4","start_station_id")\
    .withColumnRenamed("_c5","end_station_id")\
    .withColumnRenamed("_c6","rider_id")\

df_trips.write \
  .format("delta") \
  .mode("overwrite") \
  .save("/delta/trips")

spark.sql("CREATE TABLE stg_trips using DELTA LOCATION '/delta/trips'")
```
